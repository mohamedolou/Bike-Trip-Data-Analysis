## Investigation Overview

In this investigation, I wanted to look at the characteristics of trips that could be used to predict the distance of ride. The main focus was on 3 variables: duration, gender and age.

## Dataset Overview

This dataset includes information about 183412 individuals trips made in a bike-sharing 
system covering the greater San Francisco Bay Area. We have some variable like start and 
end longitude and latitude, the riding duration, the start and end station id, the user 
birth date, the user gender and the user type (suscriber or customer).

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
trips = pd.read_csv('FordGoBike.csv')
```

## Distribution of duration of riding

Ater investigation we find that generally cyclists ride between 300s and 500s. It takes on average between 5min and 8min. We might assume that users generally use bicycles for getting around and not for sports.

```python
# Let look at the distribution of the trip duration
sns.distplot(trips.duration_sec, bins=60)
plt.title('Distribution of bike duration')
plt.show()
```

## Distibution of user's age

After exploration we remark that the most people that use those bike for a trip are between 26 and 36 year old.

```python
# Let look at the distribution of the Age in the dataset (histogram)
sns.distplot(trips.Age, bins=50)
plt.title('Distribution of the age of users')
plt.show()
```

## Distibution of user's gender

We also remark that most of people that use those bike are man and the women are less than a third of men. After that we observe that few people are classified in other category.

```python
# Let see the proportion of the users genders.
sns.countplot(x='member_gender', data=trips)
plt.title('Propotion of user gender')
plt.show()
```

## Age VS Distance

Most of cyclists that ride are between 26 and 36 year old and ride on average 1.25 km

```python
trips.plot.hexbin(x="Age", y="distance(km)", gridsize=15)
plt.title('Distribution between Age and distance')
plt.show()
```
